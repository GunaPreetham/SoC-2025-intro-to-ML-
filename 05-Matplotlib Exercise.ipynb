## Exercise 1

**Follow along with these steps:**<br>
**Create a figure object called fig using plt.figure()**<br>
**Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.**<br>
**Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
import matplotlib.pyplot as plt

x = range(100)
y = [i*2 for i in x]

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='blue')

ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 100, 10000)
y = x
z = y ** 2  

fig = plt.figure()
main_ax = fig.add_axes([0, 0, 1, 1])          
inset_ax = fig.add_axes([0.2, 0.5, 0.4, 0.4]) 

main_ax.plot(x, z, 'b')
main_ax.set_xlabel('X')
main_ax.set_ylabel('Z')


inset_ax.plot(x, y, 'b')
inset_ax.set_xlim(20, 22)
inset_ax.set_ylim(30, 50)
inset_ax.set_title('zoom')
inset_ax.set_xlabel('X')
inset_ax.set_ylabel('Y')

plt.show()
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 101, 1)
y = 2 * x
z = x ** 2

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

axes[0].plot(x, y, color='blue', linestyle='--', linewidth=3)

axes[1].plot(x, z, color='red', linestyle='-', linewidth=4)

plt.tight_layout()
plt.show()
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 101, 1)
y = 2 * x
z = x ** 2

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 2))

axes[0].plot(x, y, color='blue', linestyle='-', linewidth=3)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x, z, color='red', linestyle='--', linewidth=3)
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')

plt.tight_layout()
plt.show()
```
